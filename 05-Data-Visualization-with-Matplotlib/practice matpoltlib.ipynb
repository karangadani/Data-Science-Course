```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
#functional method 
plt.plot(x,y)
plt.xlabel('X label')
plt.xlabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#oo
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X labels')
axes.set_ylabel('Y labels')
axes.set_title('Ttitle')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(2,1,figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
#axes.plot(x,y)
```

```python
fig.savefig('my_picture1.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X square')
ax.plot(x,x**3,label = 'X cubes')
ax.legend(loc = (.1,.1))
```

```python
# plot appearance 

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',lw = 2,ls = '--',)
ax.set_xlim([0,2])
```
